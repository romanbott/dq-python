```python
quad = QuadraticDifferential()
quad.smplpoles = [complex(*y) for y in np.random.uniform(-5, 5, size=[9, 2])]
quad.zeros = [complex(*y) for y in np.random.uniform(-5, 5, size=[5, 2])]
# quad.dblpoles = [complex(*y) for y in np.random.uniform(-5, 5, size=[3, 2])]
quad.phase = cm.rect(1, np.random.uniform(0, 6))
```

```python
N = 200
points = np.array([complex(*y) for y in np.random.uniform(-10,10, size=[N, 2])], dtype=np.complex)
```

```python
timer = time.time()
blabla, vec_sol = calculate_ray_vectorized(points, quad, parameters={'max_time':5000, 'lim':100})
timer = time.time() - timer
print(timer)
```

```python
def for_calc():
    sols = []
    for point in points:
        sols.append(calculate_ray(point, quad))
    return sols

def plot_lines(lines, color, offset=0, where=None, bla=True, start=0):
    for num, line in enumerate(lines):
        if where is not None:
            if bla:
                if num + start not in where:
                    continue
            else:
                if num + start in where:
                    continue
        X = np.real(line)
        Y = np.imag(line) + offset
        plt.plot(X, Y, c=color)
        
def plot_shits(quad, points):
    for zero in quad.zeros:
        plt.plot(zero.real, zero.imag, marker='o', color='darkslateblue', markersize=2)
    for pole in quad.smplpoles:
        plt.plot(pole.real, pole.imag, marker='+', color='mediumseagreen', markersize=2)
    for pole in quad.dblpoles:
        plt.plot(pole.real, pole.imag, marker='x', color='red', markersize=2)
```

```python
sel = blabla[2][0]
plot_lines(vec_sol[:N],'lightgrey',where=blabla, bla=False)
plot_lines(vec_sol[N:],'darkgrey',where=blabla, bla=False, start=N)
plot_shits(quad, points)
plot_lines(vec_sol[:N],'magenta',where=blabla, bla=True)
plot_lines(vec_sol[N:],'orchid',where=blabla, bla=True, start=N)
# plot_lines([tj], 'red')
ax = plt.axis('off')
plt.axes().set_aspect('equal', 'datalim')
plt.title('Murieron {}'.format(len(blabla)));
```
